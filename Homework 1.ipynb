### Question 5
What are the advantages and disadvantages of a very fexible (versus a less fexible) approach for regression or classifcation? Under what circumstances might a more fexible approach be preferred to a lessfexible approach? When might a less fexible approach be preferred?

Less flexible approach is beneficial when:
1. You need to communicate how a model is working especially when there are complicance requirements eg. banking 
2. Your goal is Inference; you're trying to quanitfy the relationship between predictors and a response eg. advertising and sales
3. You have a smaller sample of data and you're confident in the relationship between predictors and the response eg. defects and customer service complaints 

More flexible approach
1. Your goal is prediction and it's less important to communicate how the model is working
2. The functional form between predictors and a response is unknown and your have a large sample of data eg. personalized recommendations
